```python
import pandas as pd
import glob
import os
```

```python
def get_all_csv(folder):
    csv_files = os.path.join(folder, "*.csv")
    csv_files = glob.glob(csv_files)
    return csv_files

# "Batting Stats"
```

```python
def combine_csv(folder):
    ipl_df = pd.DataFrame()
    csv_files = get_all_csv(folder)
    for csv in csv_files:
        df = pd.read_csv(csv)
        df['Year'] = csv.split("_")[-1].split(".")[0]
        ipl_df = pd.concat([ipl_df, df])
    return ipl_df
```

```python
def get_batsman_stats():
    ipl_2023_squds = pd.read_csv('IPL 2023 squads.csv')
    ipl_df = combine_csv("Batting Stats")
    ipl_df = pd.merge(ipl_2023_squds, ipl_df, left_on='player_name', right_on='Player')
    ipl_df.drop(columns=['POS', 'Player'], inplace=True)
    ipl_df.to_excel('Batsman stats.xlsx', index=False)
    return
```

```python
def get_bowler_stats():
    ipl_2023_squds = pd.read_csv('IPL 2023 squads.csv')
    ipl_df = combine_csv("Bowling Stats")
    ipl_df = pd.merge(ipl_2023_squds, ipl_df, left_on='player_name', right_on='Player')
    ipl_df.drop(columns=['POS', 'Player'], inplace=True)
    ipl_df.to_excel('Bowler stats.xlsx', index=False)
    return ipl_df
```
